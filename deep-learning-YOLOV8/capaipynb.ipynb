```python
from ultralytics import YOLO
```

```python
modelito = YOLO('/content/best.pt')
```

```python
results = modelito.predict('/content/20230906201914.jpg',save=True, imgsz=640, conf=0.36)
```

```python
result=results[0]
mascaras = result.masks
print(len(mascaras))
```

```python
mascara1 = mascaras[0]
dato1 = mascara1.data[0].cpu().numpy() #cambiar de arreglo tensor a arreglo numpy
poligono = mascara1.xy[0]
```

```python
plt.imshow(dato1, cmap='gray')
plt.show()
```

```python
dato1im=np.uint8(dato1)
```

```python
print(dato1im.shape)
```

```python
original = cv2.imread('/content/20230906201914.jpg')
imgresize = cv2.resize(original,(640,480))
enmascarada = cv2.bitwise_and(imgresize,imgresize, mask=dato1im)
plt.imshow(cv2.cvtColor(enmascarada,cv2.COLOR_BGR2RGB))
plt.show()
```
